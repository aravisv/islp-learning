```python
#drops the rows with any cell having na value
Auto_new = Auto.dropna()
Auto_new.shape
```

```python
Auto = Auto_new
idx_80 = Auto['year'] > 80
Auto[idx_80].shape
```

```python
Auto.index
Auto.columns
```

```python
Auto_re = Auto.set_index('name')
```

```python
Auto_re.columns
```

```python
Auto_re[:3]
```

```python
rows = ['buick skylark 320','plymouth satellite']
#retrieve by the newly set index - name
Auto_re.loc[rows]
```

```python
#retrieve by the row 
#rows = [0,1,2]
#Auto_re.iloc[rows]
Auto_re.iloc[[0,1,2]]
```

```python
#use lambda function inside the loc. We can use loc to retrieve Boolean / string selections
#use iloc for integer based retrievals
Auto_re.loc[lambda df: df['year'] > 80, ['weight','origin']]
```
