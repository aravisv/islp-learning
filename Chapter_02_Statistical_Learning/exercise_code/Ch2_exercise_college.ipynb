```python
# compute the total expense for each row
college['total_expense'] = college['Room.Board'] + college['Books'] + college['Personal'] + college['Expend']
```

```python
# i thought to plot the bar graph - 2 bars elite & non elite. other parameters
# but a bar wont be able to capture the high, low, means
# box plot would be better
vars_to_plot = ['acceptance_ratio','total_expense','Grad.Rate','Terminal','PhD']
# added the 'Elite' in columns section of the dataset to avoid the error
college[vars_to_plot+['Elite']].plot.box( by = 'Elite', figsize = (10,10), layout = (3,2));
```

```python
#what if i change the definition of elite - top25% instead of top10%
college['elite_top_25'] = pd.cut(college['Top25perc'],[0,50,100],labels=['No','Yes'])
college[vars_to_plot+['elite_top_25']].plot.box( by = 'elite_top_25', figsize = (10,10), layout = (3,2));
```
