```python
# scatter plot of weight vs mpg
fig, axes = subplots(figsize = (15,12), nrows = 2, ncols = 2)
plots = [
    ('mpg', 'MPG vs Weight'),
    ('acceleration', 'Acceleration vs Weight'),
    ('displacement', 'Displacement vs Weight'),
    ('horsepower', 'Horsepower vs Weight')
]

for ax, (col, title) in zip(axes.flat, plots):
    ax.scatter(Auto['weight'], Auto[col])
    ax.set_xlabel('Weight')
    ax.set_ylabel(col.capitalize())
    ax.set_title(title)

plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Increase size for better readability
plt.figure(figsize=(10, 8))

numeric_df = Auto[['mpg', 'cylinders', 'displacement', 'horsepower', 'weight', 'acceleration']]

sns.heatmap(
    numeric_df.corr(),
    annot=True,
    cmap='coolwarm',
    fmt=".2f",
    vmin=-1,
    vmax=1
)

plt.title('Correlation Heatmap')
plt.show()
```
